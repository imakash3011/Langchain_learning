```python
## Chatprompt Template

from langchain_core.prompts import ChatPromptTemplate

prompt =ChatPromptTemplate.from_messages(
    [
        ("system", "You are an expert AI engineer. Provide me answers based on the question"),
        ("user", "{input}")
    ]
)

prompt
```

```python
# This pipe sign is used to combine prompt with llm (it is chaining elements together)
chain = prompt|llm
```

```python
response = chain.invoke({"input":"can you tell me about langsmith"})
print(response)
```

```python
type(response)
```

```python
# String output parser : Its primary purpose is to take the output generated by a model (usually text) and parse it into a more structured format that is easier to work with, such as lists, dictionaries, or custom objects.

from langchain_core.output_parsers import StrOutputParser
output_parser = StrOutputParser()
chain = prompt|llm|output_parser
```

```python
response = chain.invoke({"input": "Can you tell me about langsmith"})
print(response)
```
