```python
# Ollama

# Ollama supports embedding models, making it possible to build retrieval augmented generation (RAG)
# applications that combine text prompts with existing documents or other data
```

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings = (
    OllamaEmbeddings(model = "gemma:2b") # by default it uses llama 2
)
```

```python
embeddings
```

```python
r1 = embeddings.embed_documents(
    ["Alpha is the first letter of Greek alphabet",
     "Beta is the second letter of Greek alphabet",
     ])
```

```python
# Gemma Model creates a dimension of 2048

len(r1[0])
```

```python
embeddings.embed_query("what is the second letter of Greek alphabet ")
```

```python
# Compare the embedding of above result and the 2 statement embedding provided above to see the similarity between the embeddings
```

```python
# using other embedding model  - mxbai-embed-large

embeddings = OllamaEmbeddings(model="mxbai-embed-large")
text = "This is a test document"

query_result = embeddings.embed_query(text)
```
